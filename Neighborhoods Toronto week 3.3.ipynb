# Import Requests and BeautifulSoup and assign URL

```python
import requests
page = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page,'lxml')
```

# Find rows and columns in the table

```python
My_table = soup.find('table',{'class':'wikitable'})
rows = My_table.find_all('tr')
columns = [v.text.replace('\n', '') for v in rows[0].find_all('th')]
```

```python
df = pd.DataFrame(columns=columns)
for i in range(1, len(rows)):
               tds = rows[i].find_all('td')
# Clean and append all values to the pandas DataFrame
               values = [td.text.replace('\n','') for td in tds]             
               df = df.append(pd.Series(values, index=columns), ignore_index=True)
```

```python
df[['Postcode','Borough','Neighbourhood']].head()
```

# Clean up and customize data Frame

```python
#Ignore cells with a borough that is Not assigned
df_filtered = df[(df.Borough != 'Not assigned')].reset_index(drop=True)
df_filtered
```

```python
#'Not assigned' neighborhood -> then the neighborhood will be the same as the borough
df_filtered['Neighbourhood'] = np.where(df_filtered['Neighbourhood'] == 'Not assigned', df_filtered['Borough'], df_filtered['Neighbourhood'])
```

```python
df_filtered
```

```python
#combine neighbourhoods into one row with the neighborhoods separated with a comma
df_filtered = df_filtered.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df_filtered.head()
```

# Boroughs that contain the word "Toronto"

```python
df_toronto_borrows = df_filtered[df_filtered['Borough'].str.contains("Toronto")].reset_index()
df_toronto_borrows.head()
```

```python
#get the coordinates of Toronto
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto_borrows['Latitude'], df_toronto_borrows['Longitude'], df_toronto_borrows['Borough'], df_toronto_borrows['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```
